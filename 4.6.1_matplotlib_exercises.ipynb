```python
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
x = range(-20,20)
y = [num**2 - num + 2 for num in x]

plt.plot(x,y)
plt.annotate('origin',xy=(0,0),xytext =(-2,40),
            arrowprops={'facecolor':'magenta'})
```

```python
x = range(0,100)
y = [math.sqrt(num) for num in x]
z = [-num for num in y]

plt.plot(x,y, c='b')
plt.plot(x,z,c="blue")
plt.ylabel('y')
plt.xlabel('x')
plt.title("Square Root Function")
```

```python
x = range(-10,11)
y = [num ** 3 for num in x]

plt.plot(x,y)
plt.ylabel('y')
plt.xlabel('x')
plt.title("Cubic Function")
```

```python
import numpy
x = numpy.arange(-math.pi, math.pi, .01)
y = [math.tan(num) for num in x]

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-5,5)
plt.title('Tangent Function')
```

```python
x = range(-5,6)
y = [2 ** n for n in x]

plt.plot(x,y)
plt.title("Exponential Function")
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.figure(figsize=(12,8))

x1 = range(0,100)
y1 = [math.sqrt(num) for num in x1]
z = [-num for num in y1]

plt.subplot(221)
plt.plot(x1,y1,c='tab:blue')
plt.plot(x1,z,c='tab:blue')
plt.title("Square Root Function")

x2 = range(-10,11)
y2 = [num ** 3 for num in x2]

plt.subplot(222)
plt.plot(x2,y2)
plt.title("Cubic Function")

import numpy
x3 = numpy.arange(-math.pi, math.pi, .01)
y3 = [math.tan(num) for num in x3]

plt.subplot(223)
plt.plot(x3,y3)
plt.ylim(-5,5)
plt.title('Tangent Function')

x4 = range(-5,6)
y4 = [2 ** n for n in x4]

plt.subplot(224)
plt.plot(x4,y4)
plt.title("Exponential Function")
```

```python
plt.figure(figsize=(12,8))
plt.ylim(-5,10)
plt.xlim(-2,5)

x1 = numpy.arange(0,100,.1)
y1 = [math.sqrt(num) for num in x1]
z = [-num for num in y1]

plt.plot(x1,y1,c='tab:blue',label="$\sqrt{x}$")
plt.plot(x1,z,c='tab:blue')

x2 = numpy.arange(-10,11,.1)
y2 = [num ** 3 for num in x2]

plt.plot(x2,y2,c="tab:pink",label='$x^3$')

x3 = numpy.arange(-5, 10, .01)
y3 = [math.tan(num) for num in x3]

plt.plot(x3,y3,c="tab:brown",label='$\\tanx$')

x4 = numpy.arange(-5,6,.1)
y4 = [2 ** n for n in x4]

plt.plot(x4,y4,c="tab:cyan", label ='$2^x$')

plt.legend(loc='upper right')
```
